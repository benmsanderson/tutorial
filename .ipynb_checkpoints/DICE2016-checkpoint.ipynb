```python
import DICE2016 as dice
```

```python
def runmodel():
    
    dice.TT = np.linspace(2000, 2000+dice.tstep*dice.NT, dice.NT, dtype = np.int32)
    
    dice.InitializeLabor(dice.l,dice.NT)
    dice.InitializeTFP(dice.al,dice.NT)
    dice.InitializeGrowthSigma(dice.gsig,dice.NT)
    dice.InitializeSigma(dice.sigma,dice.gsig,dice.cost1,dice.NT)
    dice.InitializeCarbonTree(dice.cumetree,dice.NT)

    dice.x_start = np.concatenate([dice.MIU_start,dice.S_start])
    dice.bnds = dice.bnds1 + dice.bnds2
    
    result = opt.minimize(dice.fOBJ, dice.x_start, args=(-1.0,dice.I,dice.K,dice.al,dice.l,dice.YGROSS,dice.sigma,dice.EIND,dice.E,dice.CCA,dice.CCATOT,dice.cumetree,dice.MAT,dice.MU,dice.ML,dice.FORC,dice.TATM,dice.TOCEAN,dice.DAMFRAC,dice.DAMAGES,dice.ABATECOST,dice.cost1,dice.MCABATE,
         dice.CPRICE,dice.YNET,dice.Y,dice.C,dice.CPC,dice.PERIODU,dice.CEMUTOTPER,dice.RI,dice.NT), method='SLSQP',bounds = tuple(dice.bnds),options={'disp': False})
    FOptimal = dice.Optimality(result.x,dice.I,dice.K,dice.al,dice.l,dice.YGROSS,dice.sigma,dice.EIND,dice.E,dice.CCA,dice.CCATOT,dice.cumetree,dice.MAT,dice.MU,dice.ML,dice.FORC,dice.TATM,dice.TOCEAN,dice.DAMFRAC,dice.DAMAGES,dice.ABATECOST,dice.cost1,dice.MCABATE,
         dice.CPRICE,dice.YNET,dice.Y,dice.C,dice.CPC,dice.PERIODU,dice.CEMUTOTPER,dice.RI,dice.NT)

    return result,FOptimal
    
```

```python
[result,FOptimal]=runmodel()
```
